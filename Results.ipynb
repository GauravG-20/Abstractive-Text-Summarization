# 1. Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 2. Loading the saved results

Loading the saved results of each model

```python
df_atten = pd.read_csv('./Attentive_results.csv')
df_PG = pd.read_csv('./PG_results.csv')
df_RL = pd.read_csv('./RL_results.csv')
```

# 3. Plotting the graphs

```python
from turtle import color


fig = plt.figure(figsize=(15,5))
x=['Rouge-1','Rouge-2','Rouge-L']
x_axis = np.arange(len(x))

plt.subplot(1,3,1)
plt.bar(x_axis-0.2,df_atten.iloc[0].tolist()[1:],width=0.2,label='Atten Model',color='green')
plt.bar(x_axis,df_PG.iloc[0].tolist()[1:],width=0.2,label="PG Model",color='orange')
plt.bar(x_axis+0.2,df_RL.iloc[0].tolist()[1:],width=0.2,label="RL Model",color='purple')
plt.xticks(x_axis,x)
plt.xlabel('Rouge Types')
plt.ylabel('Recall')
plt.title('Recall Score Comparision')
plt.legend()

plt.subplot(1,3,2)
plt.bar(x_axis-0.2,df_atten.iloc[1].tolist()[1:],width=0.2,label='Atten Model',color='green')
plt.bar(x_axis,df_PG.iloc[1].tolist()[1:],width=0.2,label="PG Model",color='orange')
plt.bar(x_axis+0.2,df_RL.iloc[1].tolist()[1:],width=0.2,label="RL Model",color='purple')
plt.xticks(x_axis,x)
plt.xlabel('Rouge Types')
plt.ylabel('Precision')
plt.title('Precision Score Comparision')
plt.legend()

plt.subplot(1,3,3)
plt.bar(x_axis-0.2,df_atten.iloc[2].tolist()[1:],width=0.2,label='Atten Model',color='green')
plt.bar(x_axis,df_PG.iloc[2].tolist()[1:],width=0.2,label="PG Model",color='orange')
plt.bar(x_axis+0.2,df_RL.iloc[2].tolist()[1:],width=0.2,label="RL Model",color='purple')
plt.xticks(x_axis,x)
plt.xlabel('Rouge Types')
plt.ylabel('F-Scores')
plt.title('F-Score Scores Comparision')
plt.legend()


plt.show()
```
